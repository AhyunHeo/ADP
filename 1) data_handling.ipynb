```python
# 결측값 제거 // inplace 하던가 변수에 할당하던가 !!
data.dropna()   # 결측값 포함된 모든 행 삭제
data.dropna(thresh=4)   # 결측값이 아닌 값이 4개보다 많은 행만 남기기
```

```python
# 결측값 대체
data.fillna(data.mean())
```

```python
# 중복행 삭제
data['petal length'].drop_duplicates()
data[['petal width','petal length']].drop_duplicates()
```

```python
# 데이터 정렬
# axis=0 행 인덱스 기준, axis=1 컬럼명 기준
# ascending = True 오름차순, False 내림차순 정렬
data.sort_index(axis=0, ascending=False, inplace=True)
data
```
