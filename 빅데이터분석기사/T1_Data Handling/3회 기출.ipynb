### 1. 2022년 데이터 중 2022년 중앙값보다 큰 값의 데이터 수

```python
import pandas as pd
df = pd.read_csv("../data/t1-data2.csv", index_col='year')
df.head()
```

```python
m = df.loc["2022년"].median()
print(sum(df.loc["2022년",:] > m))
```

### 2. 결측치 데이터(행)을 제거하고, 앞에서부터 60% 데이터만 활용해, 'f1' 컬럼 3사분위 값을 구하시오

```python
import pandas as pd
df = pd.read_csv("../data/t1-data1.csv")
df.head()
```

```python
df = df.dropna()
df = df.iloc[:int(len(df)*0.6)]
print(df['f1'].quantile(.75))
```

### 3. 결측치가 제일 큰 값의 컬럼명을 구하시오

```python
import pandas as pd
df = pd.read_csv("../data/t1-data1.csv")
df.head()
```

```python
# 방법 1
df = pd.DataFrame(df.isnull().sum(), columns=['cnt_null'])
df = df['cnt_null'].sort_values(ascending=False)
print(df.index[0])
```

```python
# 방법 2
df = pd.read_csv("../data/t1-data1.csv")

df = df.isnull().sum()
print(df.index[3])
```
